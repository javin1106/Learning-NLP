```python
words = ["eating",  "eat", "eaten", "writing", "writes", "wrote", "programming", "programmed", "finally", "finalize", "Javin"]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem("congratulations") # Disadvatage sometimes - meaning lost
```

```python
stemming.stem("Sitting")
```

### RegexpStemmer

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer("ing$|s$|e$|able$", min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")
```

### Snowball Stemmer (better accuracy)

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word+"--->"+snowball_stemmer.stem(word))
```

Similar results as PorterStemmer but difference in some:

```python
stemming.stem("fairly"), stemming.stem("sportingly")
```

```python
snowball_stemmer.stem("fairly"), snowball_stemmer.stem("sportingly"), snowball_stemmer.stem("fairy")
```

```python
snowball_stemmer.stem("goes"), stemming.stem("goes")
```

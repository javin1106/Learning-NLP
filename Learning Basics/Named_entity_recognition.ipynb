```python
sentence = "The Eiffel Tower is a famous iron lattice tower in Paris, France, built for the 1889 World's Fair by Gustave Eiffel's company, serving as a global symbol of French culture and engineering marvel that offers stunning city views and lights up beautifully at night, attracting millions of visitors annually. The Eiffel Tower stands majestically over the Parisian skyline, a testament to 19th-century engineering."
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')

nltk.ne_chunk(tag_elements).draw()
```

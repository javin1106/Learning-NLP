```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
'''
POS- 
Noun - n (by default)
verb - v
adjective - a
adverb - r

'''
```

```python
lemmatizer.lemmatize("going", pos = 'v') # Correct as 'going' is a verb
```

```python
lemmatizer.lemmatize("going", pos = 'a')
```

```python
lemmatizer.lemmatize("going", pos = 'n')
```

```python
lemmatizer.lemmatize("going", pos = 'r')
```

```python
words = ["eating",  "eat", "eaten", "writing", "writes", "wrote", "programming", "programmed", "finally", "finalize", "Javin"]
```

```python
for word in words:
    print(word+"--->"+lemmatizer.lemmatize(word, pos = 'v'))
```

```python
corpus = """Hello my name is Javin Chutani! I am 20 years old. I am; learning NLP today! Today's task is to sleep 8 hours.
"""
```

```python
corpus
```

```python
#Tokenization
#Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
# import nltk
# nltk.download('punkt_tab')
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents: # iteratable
    print(sentence)
```

```python
# Tokenization
# Paragraphs --> Words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
# sentence--> words

for sentence in documents: # iteratable
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize # Takes care of the punctuation marks like apostrophe
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer # Considers only the last fullstop of the entire corpus as a separate word, rest are included with their parent words.
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
